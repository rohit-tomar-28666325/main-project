```python
import matplotlib.pyplot as plt

# Data
classes = ['No DR', 'Mild', 'Moderate', 'Severe', 'Proliferative DR']
images = [2400, 2400, 2400, 873, 708]

# Create the bar graph
plt.figure(figsize=(10, 6))
bars = plt.bar(classes, images, color=['#FF9999', '#66B2FF', '#99FF99', '#FFCC99', '#FF66B2'])

# Add labels and title
# plt.xlabel('Classes')
plt.ylabel('Number of Images')
plt.title('Class Distribution')

# Display the values on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 10, int(yval), ha='center', va='bottom')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

models = ['DenseNet201', 'ResNet101', 'InceptionV3', 'VGG19']

# Updated metrics for each model
accuracy = [0.6443, 0.1517, 0.6533, 0.4714]
balanced_accuracy = [0.4497, 0.2367, 0.4366, 0.3884]
fscore = [0.3809, 0.1058, 0.3589, 0.2854]
kappa_score = [0.2453, 0.0134, 0.2428, 0.1302]
precision = [0.6506, 0.1414, 0.6552, 0.4706]
recall = [0.6379, 0.1161, 0.6497, 0.3404]

x = range(len(models))

plt.figure(figsize=(12, 8))

# Adjusted plot to remove loss and AUC, and increase spacing between the remaining metrics
plt.bar([i * 1 for i in x], accuracy, width=0.15, label='Accuracy')
plt.bar([i * 1 + 0.15 for i in x], balanced_accuracy, width=0.15, label='Balanced Accuracy')
plt.bar([i * 1 + 0.30 for i in x], fscore, width=0.15, label='F-score')
plt.bar([i * 1 + 0.45 for i in x], precision, width=0.15, label='Precision')
plt.bar([i * 1 + 0.60 for i in x], recall, width=0.15, label='Recall')

plt.xlabel('Models')
plt.ylabel('Performance Metrics')
plt.title('Performance Metrics for Different Models')
plt.xticks([i * 1 + 0.30 for i in x], models)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, .99999), ncol=5)
plt.tight_layout()
plt.show()
```
